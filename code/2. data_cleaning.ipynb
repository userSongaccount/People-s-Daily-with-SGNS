# 2. Chinese words pre-processing

Remove punctuation marks and numbers block by block, and token it use *jieba*

```python
import pandas as pd
import jieba
import re
from tqdm import tqdm

# 预编译正则表达式
CJK_REGEX = re.compile(r'[\u4e00-\u9fff]')

# 定义文本处理函数
def process_text(text):
    if not isinstance(text, str):
        return []
    if len(text) > 300:
        segments = [text[i:i+150] for i in range(0, len(text), 150)]
        seg_list = []
        for seg in segments:
            seg_list.extend(jieba.lcut(seg, cut_all=False))
    else:
        seg_list = jieba.lcut(text, cut_all=False)
    return [token for token in seg_list if CJK_REGEX.search(token)]

# 分块处理函数（核心优化）
def process_large_file(input_file, output_file, chunk_size=1000):
    # 读取原始文件并分块处理
    reader = pd.read_csv(input_file, chunksize=chunk_size)
    
    # 初始化输出文件，写入表头
    with open(output_file, "w", encoding="utf-8-sig") as f:
        f.write("year,month,date,page,corpus\n")
    
    # 分块处理并逐步写入
    for chunk in tqdm(reader, desc="Processing Chunks"):
        # 处理文本 (704 chuncks)
        chunk['text'] = chunk['text'].astype(str)
        chunk['corpus'] = chunk['text'].apply(process_text)
        
        # 提取需要的列
        rm_corpus_chunk = chunk[['year', 'month', 'date', 'page', 'corpus']]
        
        # 追加写入文件（避免内存堆积）
        rm_corpus_chunk.to_csv(output_file, mode='a', header=False, index=False, encoding="utf-8-sig")

# 输入输出路径
input_file = "1946-5-15 to 2003.csv"
output_file = "rm_corpus.csv"

# 执行处理
process_large_file(input_file, output_file, chunk_size=1000)
print("save to", output_file)
```

# save cutted data into local file

```python

import os
from tqdm import tqdm

# 创建目标目录（确保路径存在）
output_dir = r"D:\CODE\data set for QTA"
os.makedirs(output_dir, exist_ok=True)

def convert_and_save(source_df, filename):
    """分块处理并保存到指定目录"""
    CHUNK_SIZE = 10000
    save_path = os.path.join(output_dir, f"{filename}.csv")  # 路径修正
    
    # 写入带BOM的UTF-8表头（避免Excel乱码）
    pd.DataFrame(columns=['tokens']).to_csv(save_path, index=False, mode='w', encoding='utf-8-sig')
    
    # 分块处理（适配图片中的数据结构）
    for i in tqdm(range(0, len(source_df), CHUNK_SIZE)):
        chunk = source_df.iloc[i:i+CHUNK_SIZE].copy()
        
        # 转换格式（处理图片中的列表字符串）
        chunk['tokens'] = chunk['corpus'].apply(
            lambda x: ', '.join(eval(x)) if pd.notnull(x) else ''
        )
        
        # 追加写入（保持编码一致）
        chunk[['tokens']].to_csv(save_path, index=False, header=False, mode='a', encoding='utf-8-sig')

# 修正年份分割（基于图片中的1976/1977分界）
c1 = corpus[corpus['year'] <= 1976]
c2 = corpus[(corpus['year'] >= 1977) & (corpus['year'] <= 1989)]  # 修正年份间隙
c3 = corpus[corpus['year'] > 1989]

# 执行保存
convert_and_save(c1, 'c1_tokens')
convert_and_save(c2, 'c2_tokens')
convert_and_save(c3, 'c3_tokens')
```
